```python
import pandas as pd
jonction_budget = pd.read_csv ('Jonction_budget.csv', sep = ';')
jonction_domicile = pd.read_csv ('Jonction_domicile.csv', sep = ';')
jonction_exterieur = pd.read_csv ('Jonction_exterieur.csv', sep = ';')
df_classement = pd.read_csv ('Classement_38_journees.csv')
df_resultats = pd.read_csv ('Resultats_Matchs.csv', names=["ID", "Date", "Journée", "Club domicile", "Score domicile", "Club extérieur", "Score extérieur", "IMG"])
df_domicile = pd.read_csv ('Stats_matchs_equipe_domicile.csv')
df_exterieur = pd.read_csv ('Stats_matchs_equipe_exterieur.csv')
```

```python
df_resultats["MATCH"] = df_resultats["Club domicile"] + " vs " + df_resultats["Club extérieur"]
df_resultats=df_resultats.drop(["IMG","ID"],axis=1)
df_resultats['Journée'] = df_resultats['Journée'].str[6:]
df_resultats
```

```python
df_resultats_stats_dom=df_resultats.merge(jonction_domicile, how='left').merge(jonction_exterieur, how='left').merge(df_domicile, how='left').merge(df_exterieur, how='left')
df_resultats_stats_dom["ID CLUB-MATCH"] = df_resultats_stats_dom["Club domicile"] + df_resultats_stats_dom["Journée"]
df_resultats_stats_dom
```

```python
df_resultats_stats_ext=df_resultats.merge(jonction_domicile, how='left').merge(jonction_exterieur, how='left').merge(df_domicile, how='left').merge(df_exterieur, how='left')
df_resultats_stats_ext["ID CLUB-MATCH"] = df_resultats_stats_dom["Club extérieur"] + df_resultats_stats_ext["Journée"]
df_resultats_stats_ext
```

```python
df_resultats_stats=df_resultats_stats_dom.append(df_resultats_stats_ext)
df_resultats_stats
```

```python
df_classement_budget=df_classement.merge(jonction_budget, how='left')
df_classement_budget=df_classement_budget.drop(["BUDGET"],axis=1)
df_classement_budget
```

```python
df_classement_budget["JOURNEE"]=df_classement_budget["JOURNEE"].apply(str)
df_classement_budget["ID CLUB-MATCH"] = df_classement_budget["CLUB thesportsdb.com"] + df_classement_budget["JOURNEE"]
df_classement_budget
```

```python
df_final=df_classement_budget.merge(df_resultats_stats)
df_final=df_final.drop(["ID CLUB-MATCH"],axis=1)
```

```python
df_final.to_excel('df_final.xlsx')
```
